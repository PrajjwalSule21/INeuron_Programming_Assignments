```python
class Array:
    def __init__(self, lst):
        self.list = lst
        
    def sum_of_list(self):
        sum_of_list = 0
        try:
            for i in self.list:
                sum_of_list = sum_of_list + i
            return f'sum of elements in list --> {sum_of_list}'
        except Exception as e:
            return 'Plese provide list of numbers:',e
        
    def multiply_of_list(self):
        multuply_of_list = 1
        try:
            for i in self.list:
                multuply_of_list = multuply_of_list * i
            return f"Multiply all numbers in the list --> {multuply_of_list}"
        except Exception as e:
            return 'Plese provide list of numbers:',e
        
    def smallest_number(self):
        l = len(self.list)
        lst = self.list.copy()
        try:
            for i in range(l):
                for j in range((i+1),l):
                    if lst[i] > lst[j]:
                        l1     = lst[i]
                        lst[i] = lst[j]
                        lst[j] = l1

            return f'Smallest number in a list --> {lst[0]}'
        
        except Exception as e:
            return 'Plese provide list of numbers:',e
        
        
    def largest_number(self):
        l = len(self.list)
        lst = self.list.copy()
        try:
            for i in range(l):
                for j in range((i+1),l):
                    if lst[i] > lst[j]:
                        l1     = lst[i]
                        lst[i] = lst[j]
                        lst[j] = l1

            return f'Largest number in a list --> {lst[-1]}'
        
        except Exception as e:
            return 'Plese provide list of numbers:',e
        
    
    def second_largest_number(self):
        l = len(self.list)
        lst = self.list.copy()
        try:
            for i in range(l):
                for j in range((i+1),l):
                    if lst[i] > lst[j]:
                        l1     = lst[i]
                        lst[i] = lst[j]
                        lst[j] = l1

            return f'Second largest number in a list --> {lst[-2]}'
        
        except Exception as e:
            return 'Plese provide list of numbers:',e
        
    
    def N_largest_elements(self, num):
        
        """
        It will take one argument,
        num : integer datatype value
        """
        lst = self.list.copy()
        final_list = []
        try:
            for i in range(0, num): 
                max_num = 0
                for j in range(len(lst)):
                    if lst[j] > max_num:
                        max_num = lst[j]

                lst.remove(max_num)
                final_list.append(max_num)
                
            final_list.sort()

            return final_list
        
        except Exception as e:
            return 'Plese provide list of numbers:',e
        
        
    def print_even_numbers(self):
        even_numbers = []
        try:
            for num in self.list:
                if num%2 == 0:
                    even_numbers.append(num)
                    
            even_numbers.sort()
            return f"List of Even Numbers --> {even_numbers}"
        
        except Exception as e:
            return 'Plese provide list of numbers:',e
        
    def print_odd_numbers(self):
        odd_numbers = []
        try:
            for num in self.list:
                if num%2 == 0:
                    pass
                else:
                    odd_numbers.append(num)
                    
            odd_numbers.sort()
            return f"List of Odd Numbers --> {odd_numbers}"
        
        except Exception as e:
            return 'Plese provide list of numbers:',e
        
        
    def remove_empty_list(self):
        lst = self.list.copy()
        try:
            for itm in lst:
                if type(itm) == list and itm == []:
                        lst.remove(itm)
            return f'After removing the Empty list --> {lst}'
        
        except Exception as e:
            return 'Plese provide list of numbers:',e
    
    
    def clone_a_list(self):
        clone = []
        try:
            for itm in self.list:
                clone.append(itm)
            return f'Orignal list --> {self.list} and Clone list --> {clone}'
        
        except Exception as e:
            return 'Plese provide list of numbers:',e  
        
        
    def count_occurences_element(self, num):
        """
        It will take one argument,
        num : integer datatype value
        
        """
        return f"Occurence of {num} in list {self.list} is : {self.list.count(num)}"
        
            
            
    def __str__(self):
        return 'This class is for list, where you can perform different-different operations on List'
```
