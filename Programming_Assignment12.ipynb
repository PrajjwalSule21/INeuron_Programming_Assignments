## 1.	Write a Python program to Extract Unique values dictionary values?

```python
class MY_dictionary:
    """
    This class is for Dictionaries only
    """
    def __init__(self, dictionary):
        self.dictionary = dictionary
        
    def extract_unique_values(self):
        """
        Objective : Method for extracting the unique values from Dictionary.
    
        
        Result : It will povide the desired list of unique values.
    
        """
        unique_values = []
        try:
            for val in self.dictionary.values():
                unique_values.append(val)
                
            unique_values = set(unique_values)
            unique_values = list(unique_values)
            
            return f"Unique values of Dictionary: {unique_values}"
        
        except Exception as e:
            return e
        
        
        
    def __str__(self):
        return f"This class is for Dictionary you can perform various task on Dictionary with the help of this class"
                
            
```
